```python
from PIL import Image
import os
import matplotlib.pyplot as plt
```

```python
# Desired dimensions
width, height = 100, 100  # Adjust as needed

# Images per row
images_per_row = 4

# Collect images
resized_images = []
for filename in os.listdir(input_dir):
    if filename.lower().endswith((".jpg", ".png", ".jpeg")):  # Filter image files
        # Open the image
        image_path = os.path.join(input_dir, filename)
        image = Image.open(image_path)
        
        # Resize the image
        resized_image = image.resize((width, height))
        resized_images.append((resized_image, filename))  # Store image and name

# Plot images in rows
total_images = len(resized_images)
rows = (total_images + images_per_row - 1) // images_per_row  # Calculate number of rows

fig, axes = plt.subplots(rows, images_per_row, figsize=(15, rows * 4))  # Adjust figsize as needed
axes = axes.flatten()  # Flatten in case of multiple rows

for ax, (image, name) in zip(axes, resized_images):
    ax.imshow(image)
    ax.axis("off")  # Hide axes

# Hide any unused subplots
for ax in axes[len(resized_images):]:
    ax.axis("off")

plt.tight_layout()
plt.show()
```
